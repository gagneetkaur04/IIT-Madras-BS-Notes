### Preprocessing

```python
def tokenization(x):
  return x.split(' ')

tweet['words'] = tweet['text'].apply(lambda x: tokenization(x))
```

```python
def lower_case(x):
  return x.lower()

tweet['text'] = tweet['text'].apply(lambda x: lower_case(x))
```
