```python
X = Data.drop(['price_range'], axis = 1)
y = Data['price_range']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 23)
```

```python
X_train.shape, y_train.shape
X_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
MM = MinMaxScaler()
LE = LabelEncoder()
```

```python
preprocess = ColumnTransformer([
    ('scaling', MM, features_to_scale)
    ], remainder = 'passthrough')
```

```python
X_train_preprocessed = preprocess.fit_transform(X_train)
X_test_preprocessed = preprocess.transform(X_test)
```

```python
type(X_train_preprocessed)
```

```python
X_train_ws = X_train.to_numpy()
X_test_ws = X_test.to_numpy()
```

```python
X_train_preprocessed.shape
```

```python
y_train = LE.fit_transform(y_train)
y_test = LE.transform(y_test)
```

```python
y_train.shape, y_test.shape
```

### **Decision Tree Classifier**


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_clf = DecisionTreeClassifier(criterion = 'entropy', random_state = 24)

DT_clf.fit(X_train_preprocessed, y_train)
```

```python
print(DT_clf.score(X_train_preprocessed, y_train))
DT_clf.score(X_test_preprocessed, y_test)
```

```python
DT_clf.fit(X_train_ws, y_train)
```

```python
print(DT_clf.score(X_train_ws, y_train))
DT_clf.score(X_test_ws, y_test)
```

* Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [
    {'max_depth': range(4, 10),
     'min_samples_leaf': [4, 5, 6],
     'min_samples_split': [6, 8]
     }
]

gscv=  GridSearchCV(DecisionTreeClassifier(), cv = 3, param_grid = param_grid, n_jobs = -1)
```

```python
gscv.fit(X_train_preprocessed, y_train)
```

```python
pd.DataFrame(gscv.cv_results_)
gscv.best_params_
gscv.best_score_
```

```python
y_train_pred = DT_clf.predict(X_train_preprocessed)
y_test_pred = DT_clf.predict(X_test_preprocessed)
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score
```

```python
print(classification_report(y_train, y_train_pred))
precision_score(y_train, y_train_pred, average = 'micro')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3, metric = 'euclidean')

knn.fit(X_train_preprocessed, y_train)
knn.score(X_train_preprocessed, y_train)
```

**Ensemble Methods**

```python
from sklearn.ensemble import BaggingClassifier
```

```python
B_clf = BaggingClassifier(estimator = DecisionTreeClassifier(random_state = 42), n_estimators = 10, random_state = 42)

B_clf.fit(X_train_preprocessed, y_train)
```

```python
B_clf.score(X_train_preprocessed, y_train)
y_train_prediction = B_clf.predict(X_train_preprocessed)
precision_score(y_train, y_train_prediction, average = 'macro')
```

```python
tree = DecisionTreeClassifier(criterion = 'entropy', random_state=42) # Declare DecisionTreeClassifier separately

bagging_clf = BaggingClassifier(base_estimator=tree)

param_grid = {
    'n_estimators': [10, 20],
    'base_estimator__ccp_alpha': [0.1, 0.5],
    'base_estimator__max_depth': [1, 2],
}

bgrid = GridSearchCV(bagging_clf, param_grid=param_grid, cv=3, verbose=1, n_jobs=-1)

bgrid.fit(X_train_preprocessed, y_train)
```
